# Task 1    

```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
def sigmoid(x):
    return 1 / 1 + np.exp(-x)
def relu(x):
    return max(x, 0)
```

```python
x1 = [1.0, 2.0]
w1 = [0.5, -0.6]
b = 0.1

z = np.dot(x1, w1) + b
a = sigmoid(z)

print(f"Z : {z}")
print(f"A : {a}")
```

# TASK 2: ACTIVATION FUNCTIONS IMPLEMENTATION

```python
x = [-5, -2, -1, 0, 1, 2, 5]

relu_x = [relu(a) for a in x]
sigmoid_x = [sigmoid(a) for a in x]

plt.figure(figsize=(10, 6))
plt.plot(x, sigmoid_x, label='Sigmoid')
plt.plot(x, relu_x, label='ReLU')

plt.xticks(x)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("ReLU vs Sigmoid Activation Function")
plt.grid(True)
plt.legend()

plt.show()
```

TASK 3: FORWARD PROPAGATION (STEP-BY-STEP COMPUTATION)

```python
def hidden_layer(x):
    w = [[0.5, -0.6],
		  [0.8, 0.2]]
    b = [0.1, -0.2]
    z = np.dot(w, x) + b

    a = [relu(i) for i in z]

    print(f"Z1 : {z[0]}, A1 : {a[0]}")
    print(f"Z1 : {z[1]}, A1 : {a[1]}")

def output_layer(x):
    w = [0.7, -1.2]
    b = 0.3

    z = np.dot(w, x) + b
    a = sigmoid(z)

    print(f"Z_out = {z}")
    print(f"Final Output: {a}")

X = [1.0, 2.0]
hidden_layer(X)
output_layer(X)
```
