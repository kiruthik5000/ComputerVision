```python
from transformers import pipeline
```

Task 1

```python
reviews = [

"The service was completed within the expected timeline and met all requirements",

"There were delays in processing and the issue was not resolved properly",

"The support team provided clear guidance and handled the request efficiently",

"The system response was slow and caused inconvenience during usage",

"The overall service experience was smooth and satisfactory",

"The issue persisted even after multiple attempts to resolve it",

"The process was well organized and easy to follow",

"There were frequent interruptions affecting system performance",

"The service quality was consistent across multiple interactions",

"The resolution provided did not address the actual problem"

]
```

# Task 3

```python
sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)
results = sentiment_pipeline(reviews[:2])
```

# Task 4

```python
for i, (review, result) in enumerate(zip(reviews, results), start=1):
    print(f"\nReview {i}:")
    print(review)
    print("Sentiment Label :", result["label"])
    print("Confidence Score:", round(result["score"], 4))
```
