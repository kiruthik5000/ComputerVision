```python
# Task 1: Compute Hidden States

x = [1, 0, 2]

Wx = 0.6
Wh = 0.4
b = 0

h = 0  # Initial hidden state
hidden_states = []

for xt in x:
    h = Wx * xt + Wh * h + b
    hidden_states.append(h)

print("Hidden states =", hidden_states)


# Task 2: Compute LSTM Cell State

f_t = 0.8          # Forget gate
i_t = 0.5          # Input gate
C_tilde = 0.6      # Candidate memory
C_prev = 1         # Previous cell state

C_t = (f_t * C_prev) + (i_t * C_tilde)

print("New cell state =", C_t)


# Task 3: Create LSTM Model

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dense
from tensorflow.keras.optimizers import Adam

model = Sequential([
    Input(shape=(3, 1)),
    LSTM(5),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer=Adam(),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Task 4: Display Model Summary

print("\nModel Summary:")
model.summary()
```
