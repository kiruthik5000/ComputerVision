```python
import numpy as np
from tensorflow.keras.metrics import binary_crossentropy
```

```python
def mean_squared_error(y_true: float, y_predicted: float) -> float:
    return np.mean(np.square(y_true - y_predicted))

def binary_cross_entropy_error(y_true: float, y_predicted: float) -> float:
    return binary_crossentropy(y_true, y_predicted)
```

```python
true_val = 1
predicted_val = 0.8
print(f"Mean Squared Error (MSE): {mean_squared_error(true_val, predicted_val)}")
print(f"Binary Cross Entropy Loss: {binary_cross_entropy_error([true_val], [predicted_val])}")
```
