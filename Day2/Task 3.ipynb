```python
def gradient_descent(w: float, n: float, dw: float) -> float:
    return w - (n * dw)
```

# Task 1


```python
w, n, dw = 0.5, 0.1, 0.2
dw_1 = gradient_descent(w, n, dw)
dw_2 = gradient_descent(w, 1.0, dw)
print(f"Updated Weight: (n = 0.1) = {dw_1}")
print(f"Updated Weight: (n = 0.1) = {dw_2}")
```
