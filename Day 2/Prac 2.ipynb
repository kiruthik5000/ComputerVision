```python
import numpy as np
```

```python
def mse(y_true: float, y_pred: float) -> float:
    return  np.mean(np.square(y_true - y_pred))
def binary_cross_entropy(y_true: float, y_pred: float) -> float:
    return -(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
def gradient_descent(w: float, n: float, dw: float) -> float:
    return w - (n * dw)
```

# Task 1

```python
y_true, y_pred = 1, 0.6


mean_sq_error = mse(y_true, y_pred)
bce = binary_cross_entropy(y_true, y_pred)

print(f" Mean Squared Error = {mean_sq_error:.4f}")
print(f" BinaryCross Entrpy = {bce:.4f}")

```

# Task 2

```python
w, dw = 0.8, 0.5
n1, n2 = 0.1, 0.5

w1 = gradient_descent(w, n1, dw)
w2 = gradient_descent(w, n2, dw)

print(f"Updated Weight (n = 0.1) = {w1}")
print(f"Updated Weight (n = 0.5) = {w2}")
```
