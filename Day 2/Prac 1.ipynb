# Task 1


```python
y_true, y_pred = 0, 0.3
x, w = 1.5, 0.6
n = 0.2

dl_dw = -(y_true / y_pred) + (1 - y_true) / (1 - y_pred)
dy_dw = x
dL_dw = dl_dw * dy_dw

print(f"dL/dŷ = {dl_dw:.4f}")
print(f"dŷ/dw = {dy_dw}")
print(f"dL/dw = {dL_dw:.4f}")
```

```python
def gradient_descent(w: float, n: float, dw: float) -> float:
    return w - (n * dw)
```

# Task 2

```python
updated_w = gradient_descent(w, n, dL_dw)

print(f"Updated w = {updated_w:.4f}")
```
