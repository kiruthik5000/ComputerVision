```python
import numpy as np
```

```python
ip_features = 3
hidden_neurons = 4
output_neurons = 1
dropout_rate = 0.5

ip_h = ip_features * hidden_neurons
h_out = hidden_neurons * output_neurons

active_neurons_after_dropout = hidden_neurons * (1 - dropout_rate)

X = np.array([2, 6, 10, 14])

mean = X.mean()
var = X.var()
```

```python
print(f"Total Weights = {ip_h + h_out}")
print(f"Active Neurons After Dropout = {active_neurons_after_dropout}")
print(f"Mean: {mean}")
print(f"Variance: {var}")
```
