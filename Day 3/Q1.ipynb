```python
import numpy as np
```

```python
ip = [20, 12, 1]
```

```python
def h1(input: np.ndarray) -> np.ndarray:

    w, bais = 20, 5
    print(f"Input -> Hidden1: Weights = {w} , Biases = {bais}")
    return np.dot(input , w) + bais

def h2(input: np.ndarray) -> np.ndarray:
    w, bais = 15, 3
    print(f"Hidden1 -> Hidden2: Weights = {w} , Biases = {bais}")

    return np.dot(input, w) + bais

def output(input: np.ndarray) -> np.ndarray:
    w, bais = 3, 1
    print(f"Hidden2 -> Output: Weights = {w} , Biases = {bais}")
    return np.dot(input, w) + bais
```

```python
val = h1(ip)
val = h2(val)
val = output(val)

print(val)
```
