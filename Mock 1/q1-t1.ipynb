```python
import numpy as np
```

```python
def relu(n: np.ndarray) -> np.ndarray:
  return np.maximum(0, n)

def sigmoid(n: np.ndarray) -> np.ndarray:
  return 1 / (1 + np.exp(-n))
```

```python
z = [-2.0, 1.0, 3.5]


for j in  z:
    print(f"Z = {j:>4} -> Sigmoid = {sigmoid(j):.3f} | ReLU = {relu(j):.3f}")
```
