```python
import numpy as np

# Loss function
def loss(w):
    return (w - 3) ** 2

# Gradient
def gradient(w):
    return 2 * (w - 3)

# Parameters
w = 0.0
learning_rate = 0.1
steps = 10

losses = []

# Initial state
print(f"Step 0 : w = {w:.3f}, Loss = {loss(w):.3f}")
losses.append(loss(w))

# Gradient Descent
for step in range(1, steps + 1):
    w = w - learning_rate * gradient(w)

    current_loss = loss(w)
    losses.append(current_loss)
    if step % 5 == 0 or step == 1:
        print(f"Step {step:^3}: w = {w:.3f}, Loss = {current_loss:.3f}")
```
