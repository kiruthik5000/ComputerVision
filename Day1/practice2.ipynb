```python
import numpy as np

def relu(n: np.ndarray) -> np.ndarray:
  return np.maximum(0, n)

def sigmoid(n: np.ndarray) -> np.ndarray:
  return 1 / (1 + np.exp(-n))

def tanh(n: np.ndarray) -> np.ndarray:
  return (np.exp(n) - np.exp(-n)) / (np.exp(n) + np.exp(-n))

def predict(input: np.ndarray) -> np.ndarray:
  weights = np.array([-0.4, 0.9])
  bias = np.array([0.2])

  return np.dot(input, weights) + bias
```

```python
X = [0.5, 1.5]

z = predict(X)
print(f"z \t\t\t= {z.item()}")

print(f"sigmoid(z) \t= {sigmoid(z).item()}")
print(f"relu(z) \t= {relu(z).item()}")
print(f"tanh(z) \t= {tanh(z).item()}")
```
