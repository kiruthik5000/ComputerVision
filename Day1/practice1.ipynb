```python
import numpy as np

def relu(n: np.ndarray) -> np.ndarray:
  return np.maximum(0, n)

def sigmoid(n: np.ndarray) -> np.ndarray:
  return 1 / (1 + np.exp(-n))

def hidden_layer(input: np.ndarray) -> np.ndarray:
  weights = np.array([0.3, -0.8])
  bias = 0.2

  Z = np.dot(input, weights) + bias
  return Z, relu(Z)

def output_layer(input: np.ndarray) -> np.ndarray:
  weights = np.array([0.7])
  bias = -0.1

  Z = np.dot(input, weights) + bias
  return Z, sigmoid(Z)
```

# Task 1

```python
input = [1.5, -2.0]

z1, a1 = hidden_layer(input)

print(f"Z1 \t\t {z1.item()}")
print(f"A1 (Relu)\t {a1.item()}")

z2, a2 = output_layer(a1)

print(f"Z2 \t\t {z2.item()}")
print(f"A2 (Sigmoid)\t {a2.item()}")
```
